1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.


```python
import math

salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# Поиск среднего арифметического
sum = 0
for salary in salaries:
  sum += salary

avg = sum / len(salaries)

# Поиск среднего квадратичного отклонения
s = 0
for salary in salaries:
  s += pow(salary - avg, 2)

# Смещенная оценка дисперсии
dn = s / (len(salaries))
dd = s / (len(salaries) - 1)
s = math.sqrt(dn)


print('Среднее арифметическое:', avg)
print('Среднее квадратичное отклонение:', s)
print('Смещенная оценка дисперсии:', dn)
print('Несмещенная оценка дисперсии:', dd)
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial
```

```python
def combinations(n, k):
  return int(factorial(n) / (factorial (k) * factorial (n - k)))
```

```python
a = combinations(5, 0) * combinations(3, 2) / combinations(8, 2) * ((combinations(5, 3) * combinations(7, 1)) / combinations(12, 4))
```

```python
b = combinations(5, 1) * combinations(3, 1) / combinations(8, 2) * ((combinations (5, 2) * combinations(7, 2)) / combinations(12, 4))
```

```python
c = combinations(5, 2) * combinations(3, 0) / combinations(8, 2) * ((combinations(5, 1) * combinations(7, 3)) / combinations(12, 4))
```

```python
print(f'Вероятность того, что три мяча белые = {a + b + c}')
```
