1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.


а) Количество крестей - 52/4=13

б) Количество карт без тузов - 52-4=48


```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
c = combinations(13, 4)
print(f"Число благоприятствующих исходов = {c}")
o = combinations(52, 4)
print(f"Общее число исходов = {o}")
print(f"Вероятность извлечь 4 крести = {c / o * 100} % ")
a = combinations(48, 4)
t = o - a
print(f"Число благоприятствующих исходов = {t} ")
print(f"Вероятность извлечь хотя бы одного туза = {t / o * 100} % ")
```

2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
l = combinations(10, 3)
print(f"Вероятность открыть дверь с первой попытки:  {1 / l * 100} %")
```
